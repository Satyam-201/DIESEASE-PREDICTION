```python
import numpy as np
import pandas as pd
import pickle
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
parkinsons_dataset = pd.read_csv("../DATASETS/parkinsons.csv")
```

```python
print(parkinsons_dataset.head())
print(parkinsons_dataset.info())
```

```python
x = parkinsons_dataset.drop(columns=['name', 'status'], axis=1)
y = parkinsons_dataset['status']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model = svm.SVC(kernel='linear')
model.fit(x_train, y_train)
```

```python
x_train_prediction = model.predict(x_train)
x_train_accuracy = accuracy_score(y_train, x_train_prediction)
print('Training Accuracy:', x_train_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
x_test_accuracy = accuracy_score(y_test, x_test_prediction)
print('Test Accuracy:', x_test_accuracy)
```

```python
pickle.dump(model, open("parkinsons_model.sav", "wb"))
pickle.dump(scaler, open("parkinsons_scaler.sav", "wb"))
```
